## ЗАДАНИЕ 1: Базовые графики matplotlib

```python
# Создайте фигуру с 4 subplots (2x2):
# 1. Линейный график функции y = x^2 * sin(x) на интервале [0, 10]
#    Добавьте заголовок, подписи осей, сетку, легенду
# 2. Гистограмма 1000 случайных точек из нормального распределения
#    Настройте количество bins=30, добавьте заголовок
# 3. Scatter plot: 50 точек (x - равномерное распределение, y = 2x + шум)
#    Настройте размер точек, прозрачность, цвет
# 4. Bar chart: сравнение 5 разных моделей по точности
#    accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
#    Добавьте значения над столбцами

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Создаем фигуру с 4 subplots (2x2)
### ваш код здесь ###
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Фигура с 4 subplots (2x2)', fontsize=16, y=1.02)

# 1. Линейный график y = x^2 * sin(x)
ax = axes[0, 0]
x = np.linspace(0, 10, 100)
y = x**2 * np.sin(x)
### ваш код здесь ###
ax.plot(x, y, label = 'y = x^2 * sin(x)')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_title('y = x^2 * sin(x)', fontsize=14)
ax.legend()
ax.grid(True)
ax.grid(color='grey', linewidth=0.5, alpha=0.3)


# 2. Гистограмма нормального распределения
ax = axes[0, 1]
data = np.random.normal(0, 1, 1000)
### ваш код здесь ###
ax.hist(data,
        bins = 30,
        alpha= 0.7,
        color='lightsteelblue',
        edgecolor='black')
ax.set_xlabel('Значение', fontsize=14)
ax.set_ylabel('Частота', fontsize=14)
ax.set_title('Гистограмма (1000 точек)', fontsize=14)
ax.grid(True)
ax.grid(color='grey', linewidth=0.5, alpha=0.3)
ax.set_axisbelow(True)

# 3. Scatter plot: y = 2x + шум
ax = axes[1, 0]
x = np.random.uniform(0, 10, 50)
y = 2 * x + np.random.normal(0, 1, 50)
### ваш код здесь ###
ax.scatter(x, y, alpha=0.6,
           color='plum',
           edgecolors='black',
           linewidth=0.5,
           s=56)
ax.set_ylabel('y', fontsize=14)
ax.set_xlabel('x', fontsize=14)
ax.set_title('Scatter plot: y = 2x + шум', fontsize=14)
ax.grid(True)
ax.grid(color='grey', linewidth=0.5, alpha=0.3)
ax.set_axisbelow(True)
ax.yaxis.set_major_locator(MultipleLocator(5))

# 4. Bar chart: точность моделей
ax = axes[1, 1]
models = ['A', 'B', 'C', 'D', 'E']
accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
### ваш код здесь ###
bars = ax.bar(models, accuracy,
              color = ['lavender', 'plum', 'lightcyan', 'orchid', 'thistle'])
ax.set_xlabel('Модель', fontsize=14)
ax.set_ylabel('Accuracy', fontsize=14)
ax.set_title('Сравнение моделей', fontsize=14)
ax.grid(True)
ax.grid(color='grey', linewidth=0.5, alpha=0.3)
ax.set_axisbelow(True)

ax.set_ylim(0.8, 1.0)

# Добавляем значения над столбцами
for bar, acc in zip(bars, accuracy):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.005,
            f'{acc:.2f}', ha='center', fontsize=11)

plt.tight_layout()
plt.show()
```

## ЗАДАНИЕ 2: Seaborn визуализации

```python
# Используйте датасет tips из seaborn:
# import seaborn as sns
# tips = sns.load_dataset('tips')

# Создайте 3 графика:
# 1. Box plot: total_bill по дням недели (day)
#    Разделите дополнительно по времени (time) с помощью hue
# 2. Violin plot: tip в зависимости от размера компании (size)
# 3. Pair plot: выберите 4 числовых признака, раскрасьте по полу (sex)

import seaborn as sns
import matplotlib.pyplot as plt

# Загружаем датасет
tips = sns.load_dataset('tips')

# Создаем фигуру для графиков
plt.figure(figsize=(10, 12))

sns.set_style("whitegrid")

# 1. Box plot: total_bill по дням с разделением по времени
# data=tips, x='day', y='total_bill', hue='time'
### ваш код здесь ###

plt.subplot(2, 2, 1)
sns.boxplot(data=tips, x='day', y='total_bill', palette='pastel', hue='time')
plt.title('total_bill по дням и времени', fontsize = 14, y = 1.01)
plt.xlabel('День недели')
plt.ylabel('Total bill')


# 2. Violin plot: tip по размеру компании
# x='size', y='tip'
### ваш код здесь ###
plt.subplot(2, 2, 2)
sns.violinplot(data=tips, x='size', y='tip', palette='deep')
# inner='quartile' - показывает квартили внутри
plt.title('Tip по размеру компании', fontsize = 14, y = 1.01)
plt.xlabel('Размер компании')
plt.ylabel('Tip')

# 3. Pair plot: 4 числовых признака, раскраска по полу
# tips[['total_bill', 'tip', 'size', 'sex']]
# Создаем отдельную фигуру для pair plot
### ваш код здесь ###
sns.pairplot(tips[['total_bill', 'tip', 'size', 'sex']], hue = 'sex',
             palette='magma', diag_kind='hist')
plt.suptitle('Pair plot по полу',  y=1.02, fontsize = 16)
plt.show()
```
